```python
import numpy as np
#test 2
import matplotlib.pyplot as plt

# Define the sub-competencies within Technical Competency
labels = ['Coding Skills', 'System Design', 'Tech Leadership', 'Innovation']

# Data for each level within the Technical Competency
# Assuming arbitrary values for demonstration
data_sm = np.array([60, 70, 60, 65])  # Senior Manager
data_d = np.array([70, 80, 75, 80])   # Director
data_sd = np.array([80, 90, 85, 90])  # Senior Director

# Repeat the first value to close the radar chart
data_sm = np.concatenate((data_sm, [data_sm[0]]))
data_d = np.concatenate((data_d, [data_d[0]]))
data_sd = np.concatenate((data_sd, [data_sd[0]]))
labels = np.concatenate((labels, [labels[0]]))

# Number of variables we're plotting.
num_vars = len(labels)

# Compute angle each bar is centered on:
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one axe per variable + add labels
ax.set_thetagrids(np.degrees(angles), labels)

# Plot data for each level
ax.plot(angles, data_sm, label='Senior Manager', linewidth=2, linestyle='solid')
ax.fill(angles, data_sm, alpha=0.1)
ax.plot(angles, data_d, label='Director', linewidth=2, linestyle='solid')
ax.fill(angles, data_d, alpha=0.1)
ax.plot(angles, data_sd, label='Senior Director', linewidth=2, linestyle='solid')
ax.fill(angles, data_sd, alpha=0.1)

# Add legend and title
plt.title('Nested Radar Chart: Technical Competency Levels', size=15)
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.05))

plt.show()
```
