```python

import matplotlib.pyplot as plt
from math import pi

# Define the categories and their respective values for each TPM level
categories = ['Program \nManagement', 'Discipline \nLeadership', 'Technical \nCompetency']
values_sm = [2, 1, 2]  # Senior Manager: Medium in Program Management and Technical Competency, Low in Leadership
values_d = [3, 2, 3]   # Director: High in Program Management and Technical Competency, Medium in Leadership
values_sd = [3, 3, 3]  # Senior Director: High in all areas
values_m = [1, 1, 1]   # Manager: Low in all areas

# Append the first value at the end of each list to close the circle
values_sm += values_sm[:1]
values_d += values_d[:1]
values_sd += values_sd[:1]
values_m += values_m[:1]

# Calculate the number of variables
num_vars = len(categories)

# Compute the angle for each category
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]  # Complete the loop

# Initialize the spider plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Draw one axis per variable and add labels
plt.xticks(angles[:-1], categories, color='black', weight='bold')

# Main sectors division with thicker lines
for label, angle in zip(ax.get_xticklabels(), angles):
    ax.plot([angle, angle], [0, 3], color='black', linewidth=3)

# Set the range for the radial axis and remove labels
plt.yticks([1, 2, 3], labels=[""]*3, color="black", size=12, weight='bold')
plt.ylim(0, 3)

# Define more contrasting colors for each role
contrasting_colors = {
    'Manager': '#0077b6',  # Light blue
    'Senior Manager': '#a3d8f6',  # Medium blue
    'Director': '#669ec4',  # Dark blue
    'Senior Director': '#3b7a9e'  # Very dark blue
}

# Plot and fill the area for each role with more contrasting colors
ax.fill(angles, values_m, color=contrasting_colors['Manager'], alpha=0.4, label='Manager')
ax.fill(angles, values_sm, color=contrasting_colors['Senior Manager'], alpha=0.4, label='Senior Manager')
ax.fill(angles, values_d, color=contrasting_colors['Director'], alpha=0.4, label='Director')
ax.fill(angles, values_sd, color=contrasting_colors['Senior Director'], alpha=0.4, label='Senior Director')

# Add plots over filled areas for clearer boundaries
ax.plot(angles, values_m, color=contrasting_colors['Manager'], linewidth=1, linestyle='solid')
ax.plot(angles, values_sm, color=contrasting_colors['Senior Manager'], linewidth=1, linestyle='solid')
ax.plot(angles, values_d, color=contrasting_colors['Director'], linewidth=1, linestyle='solid')
ax.plot(angles, values_sd, color=contrasting_colors['Senior Director'], linewidth=1, linestyle='solid')

# Add a legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Add a title
plt.title('Competency Career Growth Areas for a TPM', size=11, color="black", y=1.1)

# Display the plot
plt.show()
```
